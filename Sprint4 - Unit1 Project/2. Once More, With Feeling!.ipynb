#### Define Zillow Cleaning Function

```
def cleanZillowDf(df):

  df = df.loc[df['State'] == 'NY']
  df = pd.concat([df[['RegionName']],df.iloc[:,-36:]], axis=1)
  df = (df[df['RegionName'].isin(zipCodes)]
          .reset_index()
          .drop('index', axis=1))

  return df

def dfToCSV(df,fileName):
  df.to_csv(dataPath+fileName+'.csv', index = None, header=True)
```

#### Create cleaned sets for all zillow dfs

```
dfZil1Sub = cleanZillowDf(dfZil1Br)
dfToCSV(dfZil1Sub, 'dfZil1Sub')
print(dfZil1Sub.shape)
dfZil1Sub.head()
```

```
dfZil2Sub = cleanZillowDf(dfZil2Br)
dfToCSV(dfZil2Sub, 'dfZil2Sub')
print(dfZil2Sub.shape)
dfZil2Sub.head()
```

```
dfZil3Sub = cleanZillowDf(dfZil3Br)
dfToCSV(dfZil3Sub, 'dfZil3Sub')
print(dfZil3Sub.shape)
dfZil3Sub.head()
```

```
dfZil4Sub = cleanZillowDf(dfZil4Br)
dfToCSV(dfZil4Sub, 'dfZil4Sub')
print(dfZil4Sub.shape)
dfZil4Sub.head()
```

### NYC - (*Start with just one, ya' dummy!!!*)

#### DOB Violations!

##### Dataset info


[(documentation)](https://data.cityofnewyork.us/api/views/3h2n-5cm9/files/77115414-4d7e-4188-8547-a5ad3d10f726?download=true&filename=DD_DOB_Violations_2019-03-19.xlsx)

\

- ISN_DOB_BIS_VIOL
  - An internal code that serves as a unique value.
- BORO
  - Borough code: (1= Manhattan, 2= Bronx, 3 = Brooklyn, 4 = Queens, 5 = Staten Island)
- BIN
  - Building Identification Number assigned by Department of City Planning.		
- BLOCK
  - Tax Block assigned to the location of the proposed work by the Department of Finance.
- LOT
  - Tax Lot assigned to the location of the proposed work by the Department of Finance.
- ISSUE_DATE
  - Date that the violation was issued.	Format is YYYYMMDD	
  - Example: 20180406 = April 6, 2018
- VIOLATION_TYPE_CODE
  - A code that categorizes the violation.
  - Values are:
    - AEUHAZ1
    - B = Boiler
    - C = Construction
    - E = Elevator
    - EGNCY = Emergency
    - FISP = Façade Safety Program
    - HBLVIO = High Pressure Boiler
    - IMEGNCY = Immediate Emergency
    - LANDMK = Landmark Building
    - P = Plumbing
    - UB = Unsafe Building
    - Z = Zoning
- VIOLATION_NUMBER
  - The number associated with the violation.		
- HOUSE_NUMBER
  - The house number for the building where the violation was issued.		
- STREET
  - The street name for the building where the violation was issued.
- DISPOSITION_DATE
  - Date that the violation was dispositioned.
  - Format: YYYYMMDD
-DISPOSITION_COMMENTS
  - Additional comments by the inspector at disposition.
- DEVICE_NUMBER
  - Elevators and High Pressure Boilers are registered with the Department and issued a Device Number.
  - FISP violations may also include a device number. In this case, it is the Façade Control Number.
- DESCRIPTION
  - Notes manually entered by the DOB inspector.
- ECB_NUMBER
  - The number of an associated ECB violation, if any.
  - If there is an associated ECB Violation, you can use this number to look up the details in the DOB ECB Violations dataset.
- NUMBER
  - An ID that combines several values to create a unique ID.
  - Concatenates the following:
    - V - for violation
    - the Issue Date in MMDDYY format
    - VIOLATION_TYPE_CODE
    - VIOLATION_NUMBER

- VIOLATION_CATEGORY
  - A code and description for the current status of the violation.
  - Values are: 
    - V*-DOB VIOLATION - DISMISSED
    - V*-DOB VIOLATION - Resolved
    - V-DOB VIOLATION - ACTIVE
    - VP*-VIOLATION UNSERVED ECB- DISMISSED
    - VP-VIOLATION UNSERVED ECB-ACTIVE
    - VPW*-VIOLATION UNSERVED ECB-WORK WITHOUT PERMIT-DISMISSED
    - VPW-VIOLATION UNSERVED ECB-WORK WITHOUT PERMIT-ACTIVE
    - VW*-VIOLATION - WORK W/O PERMIT DISMISSED
    - VW-VIOLATION WORK WITHOUT PERMIT - ACTIVE
- VIOLATION_TYPE
  - A more descriptive version of the VIOLATION_TYPE_CODE field.
  - Values are:
    - AEUHAZ1-FAIL TO CERTIFY CLASS 1
    - B-BOILER
    - C-CONSTRUCTION
    - E-ELEVATOR
    - EGNCY-EMERGENCY
    - FISP-FACADE SAFETY PROGRAM
    - HBLVIO-HIGH PRESSURE BOILER
    - IMEGNCY-IMMEDIATE EMERGENCY
    - LANDMK-LANDMARK
    - P-PLUMBING
    - UB-UNSAFE BUILDINGS
    - Z-ZONING
